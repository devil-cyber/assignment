```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Conv2D,MaxPool2D,Flatten,Dense
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28,1)
x_test=x_test.reshape(x_test.shape[0],28,28,1)
```

```python
x_train=x_train/225
x_test=x_test/225
```

```python
y_train=keras.utils.to_categorical(y_train,10)
y_test=keras.utils.to_categorical(y_test,10)
```

```python
classifier=Sequential()
```

```python
classifier.add(Conv2D(6,(5,5),input_shape=(28,28,1),activation='relu'))
```

```python
classifier.add(MaxPool2D(pool_size=(2,2)))
```

```python
classifier.add(Conv2D(16,(5,5),activation='relu'))
```

```python
classifier.add(MaxPool2D(pool_size=(2,2)))
```

```python
classifier.add(Flatten())
classifier.add(Dense(120,activation='relu'))
classifier.add(Dense(84,activation='relu'))
classifier.add(Dense(10,activation='softmax'))
```

```python
classifier.compile(optimizer="adam",loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train,batch_size=128,epochs=10,validation_data=(x_test,y_test),verbose=1)
```

```python
score=classifier.evaluate(x_test,y_test)
```

```python
print('Test loss:',score[0])
print('Test acuuracy:',score[1])
```
